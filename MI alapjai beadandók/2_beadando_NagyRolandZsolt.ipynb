<p>Az adatok a <i>orders.tsv</i> szöveg fájlban vannak, ami tabulárrral elválasztott szöveges állomány.<br/><b>a)</b> Olvassa be az adatokat egy dataframe-be. Ne felejtse beállítani, hogy az elválasztó karakter most a '\t'</p>

```python
import pandas as pd
import matplotlib.pyplot as plt

#a feladat (2 pont)
ordersDF = pd.read_csv("orders.tsv", sep="\t")
ordersDF2 = ordersDF # alternatív megoldásokhoz
ordersDF
```

<p><b>b)</b> ellenőrizze a 'item_price' oszlop típusát. Alakítsa át, hogy szám legyen. <br/><i>(Segítség: 1)levágjuk a szöveg elejéről a dollár jelet, vagy 2) a dollár jelet egy üressel helyettesítjük)</i></p> 

```python
#b feladat (1 pont)

# oszlop típusa
print(ordersDF.item_price.dtype)
print(ordersDF2["item_price"].dtype)

# átalakítás 1. módszer
for sorindex in range(len(ordersDF.item_price)):
    ordersDF.item_price.iloc[sorindex, ] = ordersDF.item_price.iloc[sorindex, ][1:]
    
ordersDF["item_price"] = pd.to_numeric(ordersDF["item_price"])
print(ordersDF['item_price'][:5])


# átalakítás 2. módszer
ordersDF2["item_price"] = (ordersDF2["item_price"].replace("$", "")).astype(float)
print(ordersDF2['item_price'][:5])
```

<p><b>c)</b> Számolja ki a rendelések összegét egy 'Sum' oszlopba (az ár egységárat jelent, és quantity oszlop adja meg, hogy hány darabot rendeltek az adott ételből)</p>

```python
#c feladat (1 pont)

# 'Sum' oszlop 1. módszer  
ordersDF["Sum"] = ordersDF["item_price"] * ordersDF["quantity"]
display(ordersDF.head())

# 'Sum' oszlop 2. módszer
Sum = []
for sorindex in range(len(ordersDF2)):
    Sum.append(ordersDF2.item_price[sorindex] * ordersDF2.quantity[sorindex])

ordersDF2["Sum"] = Sum
display(ordersDF2.head())

# eredetileg a sorok tételeket jelentenek (több tétel is tartozhat ugyanahhoz a rendeléshez), tehát jelenleg a tételek összegét látjuk a 'Sum' oszlopban,
# viszont mi a rendelések összegét szeretnénk látni, ezért csoportosítjuk a tételeket rendelés szerint
# 1. módszer, eredmény egy Pandas DataFrame
display(ordersDF[["order_id","Sum"]].groupby("order_id").sum())

# 2. módszer, eredmény egy Pandas Series
ordersDF.groupby("order_id").Sum.sum()
```

<p><b>d)</b> Számítsa ki az átlagos rendelés értékét (az összes rendelés értékét kell osztani a rendelések számával)</p>

```python
#d feladat (1 pont)

# a rendelések átlagos értékét keressük (és nem a tételekét), ezért csoportosítjuk a tételeket rendelés szerint
orders_rendelésÉrtékek = ordersDF[["order_id","Sum"]].groupby("order_id").sum()

# átlag 1. módszer
print(orders_rendelésÉrtékek.Sum.sum()/len(orders_rendelésÉrtékek)) # vagy nevezőben: orders_rendelésÉrtékek.Sum.count()

# átlag 2. módszer
print(orders_rendelésÉrtékek.Sum.mean())

# a két eredmény közötti eltérés esetleg a float adattípus sajátos tárolása miatt van?
```

<p><b>h)</b> Készítsünk egy listát, hogy az egyes üdítőkből összesen mennyit rendeltek.</p>

```python
#h feladat (1 pont)

# az orders DataFrame elnevezései: üdítők -> choice_description, üdítő típusok -> item_name
# az üdítő típusok pontos nevének megkeresése kulcsszavak alapján (az adatok megismerése után feltételezzük, hogy az üdítő típusok nevében a következő kulcsszavak benne vannak)
üdítőTípusok = []
for item_name in ordersDF["item_name"].unique():
    if  (("Drink" in item_name) or ("Nectar" in item_name) or ("Canned" in item_name) or ("Bottled" in item_name) or ("Izze" in item_name)):
        üdítőTípusok.append(item_name)

print(üdítőTípusok)
# a vizet ne számítsük az üdítőTípusok közé
üdítőTípusok.remove("Bottled Water")
print(üdítőTípusok)
print()

# eredeti DataFrame szükítése csak az üdítőkre (üdítő típusok alapján)
ordersDF_csakÜdítők = ordersDF[ordersDF.item_name.isin(üdítőTípusok)]

# az üdítő típusokon belüli üdítők pontos nevének megkeresése (nem szükséges a feladathoz, csak érdekességből)
üdítők = {}
for üdítőTípus in üdítőTípusok:
    for sorindex in range(len(ordersDF_csakÜdítők)):
        if üdítőTípus in ordersDF_csakÜdítők.iloc[sorindex].item_name:
            if üdítőTípus in üdítők:
                if ordersDF_csakÜdítők.iloc[sorindex].choice_description not in üdítők[üdítőTípus]:
                    üdítők[üdítőTípus] += ordersDF_csakÜdítők.iloc[sorindex].choice_description
            else:
                    üdítők[üdítőTípus] = ordersDF_csakÜdítők.iloc[sorindex].choice_description

for üdítőTípus in üdítők:
    print(üdítőTípus, ':', üdítők[üdítőTípus])       

# például Nestea-ből(üdítő) rendeltek 6 Pack Soft Drink-ként (üdítő fajta), és Canned Soft Drink-ként (üdítő fajta) is,
# amikor az előbbiként rendelték, akkor rendelt mennyiségként 6-t számoljunk:
for sorindex, tétel in ordersDF_csakÜdítők.iterrows():
    if tétel.item_name == "6 Pack Soft Drink":
        ordersDF_csakÜdítők.loc[sorindex, ["quantity"]] = 6 * ordersDF_csakÜdítők.loc[sorindex, ["quantity"]]

rendelésSzámÜdítőnként = ordersDF_csakÜdítők.groupby("choice_description").quantity.sum() 
display(rendelésSzámÜdítőnként)

válaszListaként = []
for üdítő in rendelésSzámÜdítőnként.index.values:
     válaszListaként.append(f"{üdítő} : {rendelésSzámÜdítőnként.loc[üdítő]}")
válaszListaként
```

<h4>2. Senzorok</h4>

<p>Az <a href="http://web.uni-corvinus.hu/~fszabina/data/adls_b.txt">adls_b.txt</a> és az <a href="http://web.uni-corvinus.hu/~fszabina/data/sensors_b.txt">sensors_b.txt</a> egy kísérleti alany szokásos otthoni tevékenységeiről tartalmaz adatokat. Az adls_b.txt azt adja meg, hogy az alany mely időszakban milyen jellegű tevékenységet végzett, a sensors_b.txt pedig a házban elhelyezett szenzorok mérési adatait tartalmazza.</p>
<p>Készítsen Python nyelvű programot, amely <br/>
<b>(a)</b> a tevékenységre fordított összidő alapján csökkenő sorrendben kiírja az alany tevékenységeit! <br/>(<i>Segítség: - A tevékenységek az Activity oszlopban találhatók.<br/>
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - Az első két oszlopot dátumra kell konvertálni</i>)<br/>
<b>(b)</b> meghatározza, hogy a fürdőszobában (Bathroom) mely időintervallumokban érzékeltek a szenzorok aktivitást a 2012.11.15-ig terjedő időszakban!</p>

```python
import pandas as pd

TevékenységekDF = pd.read_csv("adls_b.txt", sep=';')
display(TevékenységekDF.head())

SzenzorokDF = pd.read_csv("sensors_b.txt", sep=';')
display(SzenzorokDF.head())
```

```python
#a feladat (1 pont)
TevékenységekDF['IdőRáfordítás'] = pd.to_datetime(TevékenységekDF['Endtime']) - pd.to_datetime(TevékenységekDF['Starttime'])

TevékenységekDF.groupby("Activity").IdőRáfordítás.sum().sort_values(ascending=False)
```

```python
#b feladat (1 pont)
SzenzorokDF[['Starttime','Endtime', 'Place']][(SzenzorokDF.Place == "Bathroom") & (SzenzorokDF.Endtime <= "2012-11-15 23:59:59")].drop_duplicates()
# drop_duplicates azért, ha esetleg a fürdőszobában 2 különböző szenzor pont ugyanakkor és ugyanaddig érzékelt volna aktivitást (csak a megfelelő időintervallumokot szeretnénk megkapni)
```
